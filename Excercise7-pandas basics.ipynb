```python
data = {'animal': ['cat', 'cat', 'snake', 'dog', 'dog', 'cat', 'snake', 'cat', 'dog', 'dog'],
        'age': [2.5, 3, 0.5, np.nan, 5, 2, 4.5, np.nan, 7, 3],
        'visits': [1, 3, 2, 3, 2, 3, 1, 1, 2, 1],
        'priority': ['yes', 'yes', 'no', 'yes', 'no', 'no', 'no', 'yes', 'no', 'no']}
labels = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j']
```

```python
#4. Creating a DataFrame df from this dictionary data which has the index labels.
df = pd.DataFrame(data, index = labels)
df
```

```python
#5. Displaying a summary of the basic information about this DataFrame and its data (hint: there is a single method that can be called on the DataFrame).
df.describe()
```

```python
#6. Returning the first 3 rows of the DataFrame df.
df.head(3)
```

```python
#7. Selecting just the 'animal' and 'age' columns from the DataFrame df.
df[['animal', 'age']]
```

```python
#8. Selecting the data in rows [3, 4, 8] and in columns ['animal', 'age'].
df[['animal', 'age']].iloc[[2,3,7]]
```

```python
#9. Selecting only the rows where the number of visits is greater than or equal to 3.
df[df['visits'] >= 3]
```

```python
#10. Checking for missing values in the data.
df.isnull().sum()
```

```python
#11. Selecting the rows where the animal is a cat and the age is less than or equal to 3.
df[(df['animal'] == 'cat') & (df['age'] < 3)]
```

```python
#12. Selecting the rows the age is between 2 and 4 (inclusive).
df[(df['age']<= 4) & (df['age'] >= 2)]
```

```python
#13. Changing the age in row 'f' to 1.5.
df['age'].loc['f'] = 1.5
df
```

```python
#14. Calculating the sum of all visits in df (i.e. find the total number of visits).
df['visits'].sum()
```

```python
#15. Calculating the mean age for each different animal in df. Explore the groupby function.
a = df.groupby('age').mean()
a
```

```python
#16. Appending a new row 'k' to df with your choice of values for each column. Then delete that row to return the original DataFrame.
df.loc['k'] = ['wolf', 10, 3, 'yes']
```

```python
# Adding the new row
df 
```

```python
# deleting the new row
df.drop('k', inplace = True) 
```

```python
#17. Counting the number of each type of animal in df.
types = df.groupby('animal').sum()
types
```

```python
#18. Sorting df first by the values in the 'age' in decending order, then by the value in the 'visit' column in ascending order (so row i should be first, and row d should be last).
df.sort_values(by = ['age', 'visits'], ascending = [False,True])
```

```python
#19. The 'priority' column contains the values 'yes' and 'no'. Replace this column with a column of boolean values: 'yes' should be True and 'no' should be False.
df['priority'].replace('yes', 1, inplace = True)
df['priority'].replace('no', 0, inplace = True)
df
```

```python
#20. In the 'animal' column, change the 'snake' entries to 'python'.
df.replace('snake', 'python', inplace = True)
df
```
